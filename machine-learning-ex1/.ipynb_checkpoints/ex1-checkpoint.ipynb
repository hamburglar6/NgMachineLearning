```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

### Import and plot data

```python
data = np.genfromtxt('./ex1/ex1data1.txt', delimiter=',')
population = data[:,0]
profit = data[:,1]

plt.figure()
plt.xlabel('Population in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.scatter(population, profit, color='red', marker='x')
plt.show()
```

### Compute the cost function

```python
theta = np.array([0.,0.]).reshape(1,2)

def compute_cost(features, target, theta):
    cost = (1/(2*target.size))*np.sum((theta[0,0]*np.ones(target.size) + theta[0,1]*features - target)**2)
    return cost

print(compute_cost(population, profit, theta))

cost_array = np.zeros(10000).reshape(100,100)
for i in range(100):
    for j in range(100):
        cost_array[i,j] = compute_cost(population, profit, np.array([(i-50)/5,(j-50)/5]).reshape(1,2))
        
plt.figure()
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.contour(np.linspace(-10,10,100), np.linspace(-10,10,100), cost_array.transpose(), 15*np.arange(10)**3)
plt.xticks(np.arange(-10, 11,2))
plt.yticks(np.arange(-10,11,2))
plt.show()
```

### Implement gradient descent

The following code doesn't actually work - it blows up for seemingly any value of $\alpha$ given enough iterations.   

```python
theta = np.array([-0.,0.]).reshape(1,2)
theta_temp = np.array([0.,0.]).reshape(1,2)
alpha = .0001
iterations = 1500
cost_check = np.zeros(iterations)
X = np.c_[np.ones(population.size), population]

for i in range(iterations):
    theta[0,0] -= alpha * np.mean(np.dot(theta_temp, X.transpose()).transpose() - profit)
    theta[0,1] -= alpha * np.mean((np.dot(theta_temp, X.transpose()).transpose() - profit) * population)
    cost_check[i] = compute_cost(population, profit, theta)
    theta_temp[0,0] = theta[0,0]
    theta_temp[0,1] = theta[0,1]
    
plt.figure()
plt.xlabel('Iteration number')
plt.ylabel(r'$J(\theta)$')
plt.plot(range(iterations), cost_check, color = 'blue')
plt.show()
```

```python
plt.figure()
plt.xlabel('Population in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.scatter(population, profit, color='red', marker='x')
plt.plot(np.linspace(0,30,50), theta[0,0] + theta[0,1]*np.linspace(0,30,50), color = 'green')
plt.show()
```

### Double checking with a black box

Here I'll use scklearn.linear_model to verify it gives the same answer as the assignment pdf.  

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(population.reshape(population.size,1), profit.reshape(population.size,1))
y_model = linear_model.predict(np.arange(0,30,.1).reshape(300,1))

plt.figure()
plt.xlabel('Population in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.scatter(population, profit, color='red', marker='x')
plt.plot(np.linspace(0,30,50), theta[0,0] + theta[0,1]*np.linspace(0,30,50), color = 'green')
plt.plot(np.arange(0,30,.1), y_model, color = 'purple', linestyle='--')
plt.show()
print(r'theta_0$ = {}'.format(linear_model.intercept_[0]))
print(r'theta_1$ = {}'.format(linear_model.coef_[0,0]))
```

The model is (sort of) doing something sensible, but the boxed implementation from scikit-learn does much better.
